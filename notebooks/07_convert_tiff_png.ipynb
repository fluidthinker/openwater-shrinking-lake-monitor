```python
from pathlib import Path
import numpy as np
import rasterio
import matplotlib.pyplot as plt
```

```python
repo_root = Path(r"C:\GIS\my_water_projects\openwater-shrinking-lake-monitor")
img_dir = repo_root / "outputs" / "images"
```

```python
def tif_to_png(tif_path: Path, png_path: Path) -> None:
    with rasterio.open(tif_path) as src:
        arr = src.read()  # (bands, rows, cols)

    rgb = np.transpose(arr, (1, 2, 0))  # (rows, cols, bands)

    # visualize() exports are usually uint8 already, but let's be safe
    if rgb.dtype != np.uint8:
        rgb = np.clip(rgb, 0, 255).astype(np.uint8)

    plt.figure()
    plt.imshow(rgb)
    plt.axis("off")
    plt.tight_layout(pad=0)
    plt.savefig(png_path, dpi=200, bbox_inches="tight", pad_inches=0)
    plt.close()
```

```python
tif_to_png(img_dir / "s2_rgb_2019_09.tif", img_dir / "s2_rgb_2019-09.png")
tif_to_png(img_dir / "s2_rgb_2025_09.tif", img_dir / "s2_rgb_2025-09.png")

print("✅ PNGs written to:", img_dir)
```
